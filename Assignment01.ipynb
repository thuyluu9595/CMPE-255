```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("used_cars_data.csv")
df
```

# Question 1 Data cleaning

```python
# Removing columns: 'S.No' and 'New_Price'
df.drop(columns=['S.No.', 'New_Price'], inplace= True)
df
```

```python
# Adding a feature named 'Car_Age'
cur_year = 2025
df['Car_Age'] = cur_year - df['Year']
df
```

```python
# Filling missing values for electric vehicle (EV)
# Identify missing Mileage for EV 
df.query("Fuel_Type == 'Electric'")
```

```python
# Driving range of Mahindra E Verito D4 is 140 km and battery capacity is 18.55kWh. 
# Source: https://www.carwale.com/mahindra-cars/verito/e-d4-2016-2019/
df.loc[4446, 'Mileage'] = "140 kmpl"

# Driving range of Toyota Prius Z4 is 23.91 kmpl. 
# Source: https://www.cardekho.com/toyota/prius-2009-2016/mileage
df.loc[4904, 'Mileage'] = "23.91 kmpl"
df.query("Fuel_Type == 'Electric'")
```

### Transform Mileage column

```python
df['Mileage'].str.extract(r'(\D+)$')[0].value_counts()
```

```python
# In the Mileage column, because kmpl is the dominant unit, all values will be converted to kmpl
# Extract number from Mileage and put in a new column named Mileage_Value
df['Mileage_Value'] = df['Mileage'].str.extract(r'(\d+\.?\d*)').astype(float)

```
