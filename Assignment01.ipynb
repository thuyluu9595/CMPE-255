# Question 2 Transformation

```python
df.info()
```

### Transform Mileage column

```python
df['Mileage'].str.extract(r'(\D+)$')[0].value_counts()
```

```python
# In the Mileage column, because kmpl is the dominant unit, all values will be converted to kmpl
# Extract number from Mileage and put in a new column named Mileage_Value
df['Mileage_Value'] = df['Mileage'].str.extract(r'(\d+\.?\d*)')[0].astype(float)
# Extract unit into a new column named Mileage_Unit
df['Mileage_Unit'] = df['Mileage'].str.extract(r'(\D+)$').astype(str)[0].str.strip()
# Conversion dictionary
rate = {
    'kmpl': 1,
    'km/kg': 1.4
}
# Convert all values to kmpl
df['Mileage_kmpl'] = df.apply(lambda row: row['Mileage_Value'] * rate.get(row['Mileage_Unit'], 1), axis=1)

# Drop all created columns and rename Mileage_kmpl to Mileage
df.drop(columns=['Mileage', 'Mileage_Value', 'Mileage_Unit'], inplace=True)
df.rename(columns={'Mileage_kmpl': 'Mileage'}, inplace=True)
df.head(30)
```

### Transform Engine Column

```python
df['Engine'].str.extract(r'(\D+)$')[0].value_counts()
```

```python
# Because there is only one unit in Engine column, unit conversion is not needed for Engine column
df['Engine'] = df['Engine'].str.extract(r'(\d+\.?\d*)')[0]
df['Engine'] = pd.to_numeric(df['Engine'], errors='coerce').astype('Int64')
df.head(20)
```

### Transform Power column

```python
df['Power'].str.extract(r'(\D+)$')[0].value_counts()
```

```python
# Since there is only one unit, which is bhp, no unit conversion needed for Power column
df['Power'] = df['Power'].str.extract(r'(\d+\.?\d*)')[0]
df['Power'] = pd.to_numeric(df['Power'], errors='coerce').astype(float)
```
